```python
# Python 2 & 3 Compatibility
from __future__ import print_function, division

# Necessary imports
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
import patsy

import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import RidgeCV
%matplotlib inline
```

```python

attr_df = pd.DataFrame(columns=['Year', 'Month', 'Term', 'Amount'])
attr_df.head()
```

```python
#import data
google_trend_term = ['SeattleRent', 'SeattleHousing', 'seattleApartment']
dataset = [None]*len(google_trend_term)
for index, term in enumerate(google_trend_term):
    dataset[index]=pd.read_csv('data//GoogleTrend/multiTimeline-{0}.csv'.format(term), header=None)
    for index, row in dataset[index].iterrows():
        split_str = row[0].split('-')
        new_row = row[:]
        new_row['Year'] = split_str[0]
        new_row['Month'] = split_str[1]
        new_row['Term'] = term
        new_row['Amount'] = row[1]
        attr_df = attr_df.append(new_row)
attr_df = attr_df.drop([0, 1], 1)
```
